```python
#Dependencies
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
import keras.backend as k

np.random.seed(12)
```

```python
#Extra Dependencies for visualizations
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#to change backend
def change_backend(convert='tensorflow'):
    
    assert convert in {'theano', 'tensorflow', 'cntk'},"Unable to change into provided backend"
    try:
        if k.backend() != convert:
            k._BACKEND=convert
            print("Backend change into : %5s"%convert)
            return
        print("Already in given backend")
           
    except Exception as error:
        print("Unable to Convert ")
        return
        
change_backend("tensorflow")
```

```python
# DataSet
size=10000
data     =    np.random.random((size,100))           #X
labels   =    np.random.randint(2,size=(size,1))     #Y
```

```python
#Sequential  Phase
model=Sequential()
model.add( Dense( 32, activation="sigmoid",
                 input_dim=100))
model.add( Dense( 1, activation="sigmoid"))

#Compilation Phase
model.compile( optimizer = "rmsprop",
                 loss = "binary_crossentropy", metrics=["accuracy"])

#Execution Phase
info=model.fit( data, labels, epochs = 10, batch_size = 32, verbose = 2)
```

```python
prediction = model.predict( data )
print(prediction)
```
