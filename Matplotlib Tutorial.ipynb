```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Defining data--- y and z data-points wrt to x
x=np.linspace(0,2.0*np.pi,101)
y=np.sin(x)
z=np.cos(x)
```

```python
"""Simple curve graph"""
#scaling of x,y axis values
xnum=np.linspace(0,7,15)
ynum=np.linspace(-1,1,11)
plt.plot(x,y,color='y',label='sin') #label can be used along with legend to show pictorially curve color with name
plt.title("Plot of some Trignometric functions")
plt.show()
```

```python
plt.plot(x,y,color='y',label='sin')
plt.plot(x,z,color='r',label='cos')
plt.xlabel('Angle in Radians')
plt.ylabel('Magnitude')
plt.title("Plot of some Trignometric functions")
plt.xticks(xnum)
plt.yticks(ynum)
plt.legend(['sin','cos']) #shows name of curve with its color or design pattern follow
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import * 
```

```python
#creating data-points
x=np.linspace(0,2.0*np.pi,101)  #x.shape= (101,)
y=np.sin(x)                     #y.shape= (101,)
z=np.cos(x)                     #z.shape= (101,)
```

```python
#ploting graph
```

```python
plt.plot(x,y,x,z) #ploting wrt x
```

```python
#drawing graph with colors
plt.xticks(xnum,color='g')
plt.yticks(ynum,color='g')
plt.xlabel("Radians",color='y')
plt.ylabel("Magnitude",color='y')
plt.title("cos and sin curve",color='r')
plt.plot(x,z,'r',x,y,'b')#ploting wrt to x
plt.grid()
plt.axis([0,6.5,-1,1])
plt.legend(['cos','sin'])
```

```python
#graph with same x-axis and diff y-axis
#seperate the figure object and axes object
#from the plotting object

#Thats means i can have diff. curves (diff axis) in same plot and diff. labels
fig,ax1=plt.subplots()  #fig shows empty graph on execution

#duplicate the axes with a different y-axis
ax2=ax1.twinx()   #ax2 and ax1 will have common x-axis and different y axis

#plot the curves on ax1,ax2 and get the curve handles
curve1,=ax1.plot(x,y,label='sin',color='r')  #Not sure how curve1, means but it make error get_color rectify
curve2,=ax2.plot(x,z,label='cos',color='b')
#Make curve list to acess parameters
curves=[curve1,curve2]

#x-axis labels via axes
ax1.set_xlabel("Magnitude",color=curve1.get_color()) #because we have same x-axis we will get only 1 label not 2

#y-axis lables via axes
ax1.set_ylabel("Angle/values",color=curve1.get_color()) #we have 2 diff. y -axis so need to draw 2
ax2.set_ylabel("Angle/values",color=curve1.get_color())


#y-ticks via axes
ax1.tick_params(axis='y',color=curve1.get_color()) 
ax1.tick_params(axis='y',color=curve2.get_color())

#x-ticks via axes
ax1.tick_params(axis='x',color=curve1.get_color()) #one tick is sufficient

#set x and y ticks via axes
ax1.set_xticks(xnum)
ax1.set_yticks(ynum)
ax2.set_yticks(ynum)

#to make  grids using axes 2
#ax2.grid(color=curve2.get_color())  #try this also
#ax1.grid(color=curve2.get_color())
ax2.grid(color='w') #missing pixels are actually grid for ax2 's grid
ax1.grid(color='g')

#do not display following grid
ax1.xaxis.grid("False")
#add legend via axis1 or axis 2 object
#one ax1.legend or ax2.legend is sufficient enough
ax1.legend(curves,[curve.get_label() for curve in curves])
#ax1.legend()
#ax2.legend() 

#Global figure properties
plt.title("Plot of Sine and Cos",color='g')
#plt.grid()
plt.show()
```
